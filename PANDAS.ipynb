```python
# ==================================================
# PANDAS IN PYTHON – COMPLETE NOTES WITH EXAMPLES
# ==================================================

# Pandas is a Python library used for:
# - Data analysis
# - Data cleaning
# - Data manipulation
# - Working with tables (rows & columns)

import pandas as pd

# ==================================================
# 1. CORE DATA STRUCTURES
# ==================================================

# Pandas mainly works with:
# 1. Series  -> 1D data
# 2. DataFrame -> 2D data (table)

# ==================================================
# 2. SERIES
# ==================================================

# A Series is a one-dimensional array with labels (index)

s = pd.Series([10, 20, 30, 40])

# Access values
# s[0] -> 10

# Series with custom index
s2 = pd.Series([100, 200], index=["A", "B"])

# ==================================================
# 3. DATAFRAME
# ==================================================

# A DataFrame is a table with rows and columns

data = {
    "Name": ["Amit", "Rahul", "Neha"],
    "Age": [25, 30, 28],
    "Salary": [50000, 60000, 55000]
}

df = pd.DataFrame(data)

# ==================================================
# 4. BASIC DATAFRAME FUNCTIONS
# ==================================================

# View first rows
df.head()

# View last rows
df.tail()

# Shape (rows, columns)
df.shape

# Column names
df.columns

# Data types
df.dtypes

# Summary info
df.info()

# Statistical summary
df.describe()

# ==================================================
# 5. COLUMN OPERATIONS
# ==================================================

# Select single column
df["Age"]

# Select multiple columns
df[["Name", "Salary"]]

# Add new column
df["Bonus"] = df["Salary"] * 0.10

# Rename column
df.rename(columns={"Salary": "Income"}, inplace=True)

# Drop column
df.drop("Bonus", axis=1, inplace=True)

# ==================================================
# 6. ROW OPERATIONS
# ==================================================

# Select row by index
df.loc[0]

# Select rows by condition
df[df["Age"] > 26]

# Select by index position
df.iloc[1]

# ==================================================
# 7. SORTING DATA
# ==================================================

# Sort by column
df.sort_values("Age")

# Sort descending
df.sort_values("Age", ascending=False)

# ==================================================
# 8. FILTERING DATA
# ==================================================

# Single condition
df[df["Age"] > 25]

# Multiple conditions
df[(df["Age"] > 25) & (df["Income"] > 50000)]

# ==================================================
# 9. HANDLING MISSING VALUES
# ==================================================

# Check missing values
df.isnull()

# Count missing values
df.isnull().sum()

# Fill missing values
df.fillna(0)

# Drop rows with missing values
df.dropna()

# ==================================================
# 10. APPLY FUNCTION
# ==================================================

# Apply function to column
df["Age"] = df["Age"].apply(lambda x: x + 1)

# ==================================================
# 11. VALUE COUNTS
# ==================================================

df["Age"].value_counts()

# ==================================================
# 12. GROUPBY (VERY IMPORTANT)
# ==================================================

# Group data and aggregate
df.groupby("Age")["Income"].mean()

# Multiple aggregations
df.groupby("Age").agg({
    "Income": "mean",
    "Name": "count"
})

# ==================================================
# 13. MERGE DATAFRAMES
# ==================================================

df1 = pd.DataFrame({
    "id": [1, 2],
    "name": ["A", "B"]
})

df2 = pd.DataFrame({
    "id": [1, 2],
    "salary": [50000, 60000]
})

# Merge (like SQL join)
pd.merge(df1, df2, on="id")

# ==================================================
# 14. CONCAT DATAFRAMES
# ==================================================

pd.concat([df1, df2], axis=0)

# ==================================================
# 15. READ FILES
# ==================================================

# Read CSV file
# df = pd.read_csv("data.csv")

# Read Excel file
# df = pd.read_excel("data.xlsx")

# ==================================================
# 16. WRITE FILES
# ==================================================

# Write to CSV
# df.to_csv("output.csv", index=False)

# Write to Excel
# df.to_excel("output.xlsx", index=False)

# ==================================================
# 17. STRING FUNCTIONS
# ==================================================

# Convert to uppercase
df["Name"].str.upper()

# Check contains
df["Name"].str.contains("A")

# ==================================================
# 18. DATE & TIME
# ==================================================

# Convert column to datetime
# df["Date"] = pd.to_datetime(df["Date"])

# Extract year
# df["Year"] = df["Date"].dt.year

# ==================================================
# 19. DUPLICATES
# ==================================================

# Check duplicates
df.duplicated()

# Drop duplicates
df.drop_duplicates()

# ==================================================
# 20. RESET & SET INDEX
# ==================================================

df.reset_index(drop=True)

df.set_index("Name", inplace=True)

# ==================================================
# 21. IMPORTANT INTERVIEW POINTS
# ==================================================

# DataFrame is mutable
# Pandas is built on NumPy
# groupby is very important
# apply + lambda is common
# Pandas handles missing data easily
```
