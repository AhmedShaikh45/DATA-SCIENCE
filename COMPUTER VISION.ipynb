```python
#Rotating the image

import cv2

img = cv2.imread("photo.png")

(h, w) = img.shape[:2]   # ✅ correct
center = (w // 2, h // 2)  # (x, y)

M = cv2.getRotationMatrix2D(center, 90, 1.0)
rotated = cv2.warpAffine(img, M, (w, h))

cv2.imshow("Original Image", img)
cv2.imshow("Rotated Image", rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Shape Detection
import cv2

# Read image
img = cv2.imread('shape.png')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Threshold
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Find contours
contours, hierarchy = cv2.findContours(
    thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE
)

for contour in contours:
    # Approximate contour
    epsilon = 0.01 * cv2.arcLength(contour, True)
    approx = cv2.approxPolyDP(contour, epsilon, True)

    corners = len(approx)

    # Shape detection
    if corners == 3:
        shape_name = 'Triangle'
    elif corners == 4:
        shape_name = 'Rectangle'
    elif corners == 5:
        shape_name = 'Pentagon'
    elif corners > 5:
        shape_name = 'Circle'
    else:
        shape_name = 'Unknown'

    # Draw contour
    cv2.drawContours(img, [approx], 0, (0, 255, 0), 2)

    # Put text
    x, y = approx.ravel()[0], approx.ravel()[1] - 10
    cv2.putText(
        img, shape_name, (x, y),
        cv2.FONT_HERSHEY_COMPLEX, 0.6, (255, 0, 0), 2
    )

# Show output
cv2.imshow('Contours Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Real Time Face Detection
import cv2

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 5)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    cv2.imshow('Web Cam', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# Face, Smile, Eye Detection
import cv2

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
smile_cascade = cv2.CascadeClassifier('haarcascade_smile.xml')

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 5)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

        roi_gray = gray[y:y + h, x:x + w]
        roi_color = frame[y:y + h, x:x + w]

        eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 10)
        if len(eyes) > 0:
            cv2.putText(frame, 'Eyes Detected', (x, y - 30),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

        smile = smile_cascade.detectMultiScale(roi_gray, 1.7, 20)
        if len(smile) > 0:
            cv2.putText(frame, 'Smiling', (x, y - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    cv2.imshow('Smart Face Detector', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#WEB CAM
import cv2

cap = cv2.VideoCapture(0)  # 0 = default webcam

if not cap.isOpened():
    print("Error: Could not access webcam")
    exit()

paused = False

while True:
    if not paused:
        success, frame = cap.read()
        if not success:
            break
        cv2.imshow("Webcam", frame)

    key = cv2.waitKey(1) & 0xFF

    # Pause / Resume
    if key == ord('p'):
        paused = not paused

    # Quit
    elif key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#Dilation
import cv2
import numpy as np

img = cv2.imread("ahmed.png", 0)  # grayscale

# Convert to binary (important for dilation)
_, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Create kernel
kernel = np.ones((5, 5), np.uint8)

# Apply dilation
dilated = cv2.dilate(binary, kernel, iterations=1)

cv2.imshow("Original", binary)
cv2.imshow("Dilated", dilated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Erode
import cv2
import numpy as np

# Read image in grayscale
img = cv2.imread("ahmed.png", 0)

# Convert to binary
_, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Create kernel
kernel = np.ones((5, 5), np.uint8)

# Apply erosion
eroded = cv2.erode(binary, kernel, iterations=1)

cv2.imshow("Original Binary", binary)
cv2.imshow("Eroded Image", eroded)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Dilation vs Erode
import cv2
import numpy as np

img = cv2.imread("ahmed.png", 0)

# Convert to binary
_, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Kernel
kernel = np.ones((5,5), np.uint8)

# Dilation
dilated = cv2.dilate(binary, kernel, iterations=1)

# Erosion
eroded = cv2.erode(binary, kernel, iterations=1)

cv2.imshow("Original Binary", binary)
cv2.imshow("Dilated", dilated)
cv2.imshow("Eroded", eroded)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
