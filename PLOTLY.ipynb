BUBBLE CHART

```python
import numpy as np
```

```python
x1,y1 = np.random.randint(1,100,10),np.random.randint(1,100,10)
x2,y2 = np.random.randint(1,100,10),np.random.randint(1,100,10)
x3,y3 = np.random.randint(1,100,10),np.random.randint(1,100,10)

```

```python
data=[]
i = 2021
for x,y in [(x1,y1),(x2,y2),(x3,y3)]:
    name="year  -"+str(i)
    i+=1
    trace=go.Scatter(x=x,y=y,mode="markers",marker_size=np.random.randint(20,50,10),name=name)
    data.append(trace)
go.Figure(data,layout=go.Layout(title="BUBBLE CHART BY CATEGORY"))
```
