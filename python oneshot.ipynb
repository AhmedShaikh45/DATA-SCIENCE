#### Python Data Types – One-Page Exam Notes

Python data types define the type of data a variable can store. They decide how data is stored and manipulated in memory.

---

#### 1. Numeric Data Types
Used to store numbers.

- **int** → Whole numbers (positive/negative)  
  Example: `10`, `-200`

- **float** → Decimal numbers  
  Example: `1.23`, `3.45`

- **complex** → Real + Imaginary numbers (`j`)  
  Example: `3 + 5j`

---

#### 2. Boolean & NoneType

- **bool** → Logical values  
  Values: `True`, `False`

- **NoneType** → Represents absence of value  
  Value: `None`

---

#### 3. Sequence Data Types
Ordered collections.

- **string (str)**  
  - Collection of characters  
  - Immutable  
  Example: `"Python"`

- **list**  
  - Ordered collection  
  - Mutable  
  - Allows mixed data types  
  Example: `[1, 2, 3]`

- **tuple**  
  - Ordered collection  
  - Immutable  
  Example: `(10, 20)`

---

#### 4. Set Data Types
Unordered collections of unique elements.

- **set**  
  - Mutable  
  - No duplicate values  
  Example: `{1, 2, 3}`

- **frozenset**  
  - Immutable version of set  
  Example: `frozenset([1, 2, 3])`

---

#### 5. Mapping Data Type

- **dictionary (dict)**  
  - Stores data in key–value pairs  
  - Keys must be unique & immutable  
  Example:
  ```python
  {"name": "Ahmed", "age": 22}


#### PEMDAS – One-Page Exam Notes

PEMDAS defines the **order of operations** used to evaluate mathematical and Python expressions.

---

#### PEMDAS Order (Highest → Lowest)

1. **P – Parentheses ( )**
   - Expressions inside parentheses are evaluated first
   - Nested parentheses are solved from inner to outer

2. **E – Exponents ( ** )**
   - Power and root operations
   - Evaluated after parentheses

3. **M / D – Multiplication & Division ( * , / )**
   - Same priority level
   - Solved from **left to right**

4. **A / S – Addition & Subtraction ( + , - )**
   - Same priority level
   - Solved from **left to right**

---

#### Important Rules
- Parentheses always have the highest priority
- Exponentiation is done before multiplication and division
- Multiplication and division are evaluated left to right
- Addition and subtraction are evaluated left to right
- Same-priority operations follow **left-to-right rule**

---

````python
class Profile:
    def __init__(self,name,age,grade,course):
        self.name = name
        self.age = age
        self.grade = grade
        self.course = course

    def display(self):
        print(self.name,self.age,self.grade,self.course)

p1 = Profile('ahmed',21,'A+','Comps')
p2 = Profile('shaikh',20,'B+','Mech')

print(p1.name)
print(p2.course)
p1.display()
p2.display()
````

````python
# =========================================
# POLYMORPHISM – SIMPLE EXAMPLE
# =========================================
class Animal():
    def sound(self):
        print('Animal sound')


class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

# Same method name, different behavior
d = Dog()
c = Cat()

d.sound()
c.sound()
````

````python
# =========================================
# POLYMORPHISM USING LOOP
# =========================================

class Dog:
    def sound(self):
        print("Dog barks")

class Cat:
    def sound(self):
        print("Cat meows")

class Cow:
    def sound(self):
        print("Cow moos")

animals = [Dog(), Cat(), Cow()]

for animal in animals:
    animal.sound()
````

````python
class BankAccount:

    def __init__(self,acc_no,balance):
        self.acc_no = acc_no
        self.__balance = balance

    def deposit(self,amount):
        self.__balance += amount
        print(f'Amount deposited {amount}')

    def get_balance(self):
        return self.__balance

account = BankAccount('12345',5000)

account.deposit(2000)

print(account.get_balance()) 
````

````python
#Inheritence
# =========================================
# METHOD OVERRIDING
# =========================================

class Animal:
    def sound(self):
        print("Animal sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

d = Dog()
d.sound()
````

````python
# =========================================
# SIMPLE INHERITANCE
# =========================================

class Parent:
    def show(self):
        print("This is Parent class")

class Child(Parent):
    pass

obj = Child()
obj.show()
````
